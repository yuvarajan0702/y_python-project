```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
def get_stock_percentage_change(stock_symbol):

    stock = yf.Ticker(stock_symbol)

    stock_data = stock.history(period="1y")

    high_52week = stock_data['High'].max()
    low_52week = stock_data['Low'].min()

    current_price = stock_data['Close'].iloc[-1]

    percentage_change = ((current_price - low_52week) / low_52week) * 100

    return high_52week, low_52week, current_price, percentage_change
```

```python
stock_symbols = ['TCS.NS', 'RELIANCE.NS', 'INFY.NS', 'HDFCBANK.NS', 'ICICIBANK.NS']
```

```python
results = []

for stock in stock_symbols:
    high_52week, low_52week, current_price, percentage_change = get_stock_percentage_change(stock)
    results.append([stock, high_52week, low_52week, current_price, percentage_change])

df = pd.DataFrame(results, columns=['Stock', '52 Week High', '52 Week Low', 'Current Price', '% Change from Low'])
df
```
